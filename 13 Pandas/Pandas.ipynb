###### Pandas Data Structures.
- There are two core objects in pandas: the <b>DataFrame</b> and the <b>Series.</b>
- A <b>Dataframe</b> is two dimensional data structure that is aligned in tabular format i.e in row and columns.
- A <b>Series</b>, by contrast, is a sequence of data values. If a DataFrame is a table, then a Series is a list.
- Each column in a Dataframe is a <b>Series</b>.
- Pandas also has a 3-dimensional data structure known as a <b>Panel</b>

### Selecting rows and columns

We have two options:

- .loc - locates by row and column label.
- .iloc- locates by row and column index.

#### The main distinction between loc and iloc is:
- loc is label-based, which means that you have to specify rows and columns based on their row and column labels.
- iloc is integer position-based, so you have to specify rows and columns by their integer position values (index position).
- Read more: https://towardsdatascience.com/how-to-use-loc-and-iloc-for-selecting-data-in-pandas-bd09cb4c3d79

### Dropping the indices.
- When we select a subset of data, we end up with non consecutive indices.
- This make it hard to use iloc on that subset.
- We can reset the indices using the <b>reset_index()</b> method.
- If we want to drop them altogether, we can pass the argument <b>drop=True</b> to the reset_index() method.

```python
data = df.iloc[[10, 1, 5, 700, 60, 26], [18, 19, 20]]
data.reset_index(inplace=True, drop=True)
data
```

### Performing Column Operations.
- For instance we want to make all words in a column to uppercase.
- We can use the **apply** function to every value in a particular column.
- The **apply()** function is used to apply a function along an axis of the DataFrame. 
- Objects passed to the function are Series objects whose index is either the DataFrame's index (axis=0) or the DataFrame's columns (axis=1).
###### Using lambda function.
- Lambda function is basically shorthand for writing python function.
- Lambda function dont have a name or a return statement.
- Does not take take any arguments.

```python
df_athletes['first_name'] = df_athletes['Name'].apply(lambda x:x.split(' ')[0])
df_athletes
```

```python
def last_name(name):
    l_name = name.split(' ')[-1]
    return l_name

df_athletes['last_name'] = df_athletes['Name'].apply(last_name)

    
df_athletes
```

```python
df_athletes['age-5'] = df_athletes['Age'].apply(lambda x: x - 5)
```

### Calculating Aggregate Functions.

###### Groupby Method.

```python
#We want to get the churn for each gender.
churn_data_gender = df.groupby(['gender']).Churn.value_counts().reset_index(name='Total')
churn_data_gender
```

### Pivot tables.
- We use Pivot Tables to for reorganizing a dataframe.
- This process is called <b>Pivoting</b>
- The new table is called a <b> Pivot Table</b>

```python
#Pivot table showing the churn between male and females.
churn_data_gender_pivot=churn_data_gender.pivot(columns='Churn', index = 'gender', values= 'Total')
churn_data_gender_pivot
```

### Concatenate Dataframes 
- Sometimes a dataset is broken to multiple tables eg Data cans be split into multiple CSV files so that each download is smaller.
- We therefore need to reconstruct a single Dataframe from multiple smaller dataframes, we can use the method <b>pd.concat([df1, df2, ___ dfn]) </b> 
- This method only works if all the columns are same in all dataframes.

```python
athlete2 = {"Name": ['Joel Embiid', 'Megan Rapinoe', 'Didier Drogba'], 
            "Age": [29, 36, 43], 
            "Gender": ["Male", "Female", "Male"],
            "Game": ["Basketball", "Football", "Football"],
            "first_name": ["Joel", "Megan", "Didier"]}
df_athletes2 = pd.DataFrame(data=athlete2)
```

```python
df_athletes2
```

```python
athlete_combined = pd.concat([df_athletes, df_athletes2]).reset_index(drop=True)
```

```python
athlete_combined
```
