## Sales Analysis

- In this section we are going to to analyze and answer business questions about 12 months worth of sales data. 
- The data contains hundreds of thousands of electronics store purchase records broken down by month, product type, cost, purchase address, etc.

 In this section we explore 5 high level business questions related to our data:

* What was the best month for sales? How much was earned that month?
* What city sold the most product?
* What time should we display advertisements to maximize the likelihood of customerâ€™s buying product?
* What products are most often sold together?
* What product sold the most? Why do you think it sold the most?

```python
import pandas as pd
import os
```

### Merge 12 month of sales data
- Since data is organized in months, we need to merge all the months to one dataframe.

```python
#Start with an empty dataframe that will contain 12 month data.
all_months_data = pd.DataFrame()
files =[file for file in os.listdir('./Sales_Data')]
for file in files:
    df = pd.read_csv('./Sales_Data/' + file)
    all_months_data = pd.concat([all_months_data, df])
```

```python
df.head()
```

### Save the dataframe to CSV file.

```python
all_months_data.to_csv('all_data.csv', index=False)
```

### Read the new csv containing all data

```python
data = pd.read_csv('all_data.csv')
```

```python
data.head()
```

### Clean up the data

- You need to figure out what we need to clean.
- You find things you need to clean as you perform operations and get errors.
- Based on the error, you decide how you should go about cleaning the data.



###### Find missing values

```python
nan_df = data[data.isna().any(axis=1)]
nan_df.head()
```

```python
data.isna().any(axis=1)
```

###### Dropping rows with missing values

```python
all_data = data.dropna(how='all').reset_index(drop=True)
all_data.head()
```

```python
all_data.isna().any(axis=1)
```

###  Checking for duplicates
- Often we see duplicated rows of data in a DataFrame.
- This could happen due to erros in data collection or in saving and loading data.
- We can check for duplicates using pandas .duplicated() method.

```python
all_data.duplicated().value_counts()
```

```python
all_data.shape
```

### Dropping Duplicates 
- Duplicates always don't add any value to the dataset and can cause bias in Ml Models.
- Its good to drop the duplicate values.
- We use .drop_duplicates method.

```python
all_data.drop_duplicates(inplace=True)
```

```python
all_data.duplicated().value_counts()
```

#### Get rid of invalid dates in order date column

```python
#find text in order date 
temp_df=all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head()
```

```python
# create a dataframe without or in order date
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
all_data
```

###### Check column Type

```python
all_data.info()
```

- Price Each & Quantity Ordered Pandas read them as object instead of numbers
- We need to convert them to the corrrect data type

###### Convert columns to correct data type

```python
#make int
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])

# make float 
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data.dtypes
```

### Add sales column
- Multiply the quantity ordered by the price.

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data
```

## Data Analysis

###### 1) What was the best month for sales? How much was earned that month?

```python
#TOTAL SALES
all_data['Sales'].sum()
```

```python
monthly_sales = all_data.groupby('month').sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt
months =range(1,13)
plt.bar(months,monthly_sales['Sales']/1000000)
plt.xticks(months)
plt.ylabel('sales in usd (millions) ')
plt.xlabel('Month number ')
plt.show()
```

### 2) What city sold the most product?
###### Add city column

```python
all_data
```

```python
def get_city(address):
    return address.split(",")[1].strip(" ")

def get_state(address):
    return address.split(",")[2].split(" ")[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : f'{get_city(x)}, {get_state(x)}')
all_data
```

#### What city sold the most product?

```python
sales_by_city = all_data.groupby(['City']).sum().sort_values(by='Sales', ascending=False).reset_index(level=0)
```

```python
sales_by_city
```

```python
plt.bar(sales_by_city['City'], sales_by_city['Sales'])
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.xticks(sales_by_city['City'].values, rotation='vertical', size=8)
plt.show()
```

#### 3) What time should we display advertisements to maximize the likelihood of customerâ€™s buying product?

```python
# Add hour column  
all_data['Hour'] = all_data['Order Date'].dt.hour
# Add minute column
all_data['Minute'] = all_data['Order Date'].dt.minute
```

```python
sales_per_hr = all_data.groupby('Hour')['Order ID'].count().reset_index(level=0, name='Orders')
sales_per_hr
```

```python
plt.plot(sales_per_hr['Hour'], sales_per_hr['Orders'])
plt.xticks(sales_per_hr['Hour'])
plt.grid()
plt.show()
```

#### 4) What products are most often sold together?

###### Group product with the same order id

- Groupby the 'Order ID' and 'Product' columns, then call transform which will return data aligned to the original df
- Apply a lambda where we join the text entries:

```python
df = all_data
df['Grouped'] = all_data.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df.head()
```

#### Drop Duplicates

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

##### Count pairs that arise from combinations in your product lists.Â¶

```python
from itertools import combinations
from collections import Counter

count =Counter()

for row in df['Grouped']:
    row_list =row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

#### 5) What product sold the most? Why do you think it sold the most?

```python
sales_per_product = all_data.groupby('Product').sum()['Quantity Ordered'].reset_index(level=0, name='Quantity')
sales_per_product
```

```python
plt.bar(sales_per_product['Product'], sales_per_product['Quantity'])
plt.xticks(sales_per_product['Product'].values, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each'].reset_index(level=0, name='Price')
```

```python
prices
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax2.plot(prices['Product'], prices['Price'], color='red')
ax1.bar(sales_per_product['Product'], sales_per_product['Quantity'])

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(prices['Product'].values, rotation='vertical', size=8)
fig.show()
```

Low price items fetch more sales quantity
