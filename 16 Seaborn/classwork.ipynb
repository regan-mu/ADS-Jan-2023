```python
#load dataset tips 
tips = sns.load_dataset('tips')

# preset themes
sns.set_style('ticks')

# plot 
sns.stripplot(x="day", y="total_bill", data=tips)
sns.despine(bottom=True,left=True)
plt.show()
```

```python
%matplotlib inline
#load dataset tips 
tips = sns.load_dataset('tips')

# preset themes
sns.set_style('darkgrid')
sns.set_context('notebook')

# plot 
sns.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
x = list(range(0, 6))
y = [n**2 for n in x]

sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.8, rc={'lines.linewidth': 6, 'grid.linewidth': 3})

# plot 
sns.lineplot(x=x, y=y, marker='o', linestyle="-.", color="red")
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.show()
```

```python
# The default plot kind is a scatterplot:
# FIGURE LEVEL FUNCTION
plt.figure(figsize=(14,7))
sns.set_style('whitegrid')
sns.set_context("notebook")
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="sex")
plt.show()
```

```python
# AXES LEVEL FUNCTION
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
# The default plot kind is a histogram:
sns.displot(data=titanic, x='age', kind="kde")
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=titanic, x='age')
plt.show()
```

```python
sns.kdeplot(data=titanic, x='age')
plt.show()
```
