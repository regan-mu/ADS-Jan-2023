```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(pd.__version__)
```

```python
url = "data/dataonline.csv"
df = pd.read_csv(url)
df.head()
```

```python
df['Revenue'] = df['UnitPrice'] * df['Quantity']
```

```python
df.info()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["InvoiceMonth"] = pd.DatetimeIndex(df["InvoiceDate"]).month
df["InvoiceYear"] = df["InvoiceDate"].dt.year
df.head()
```

```python
df_sales = df.groupby('Country').sales.sum().reset_index()
```

```python
df_revenue = df.groupby(["InvoiceMonth", "InvoiceYear"])["Revenue"].sum().reset_index()
df_revenue
```

```python
df_sales.columns=['Country', 'Sales']
```

```python
df_sales.sort_values(by='Sales', inplace=True, ascending=False)
```

```python
df_sales.reset_index(inplace=True, drop=True)
top_10_sales = df_sales.iloc[-5:]
top_10_sales
```

```python
plt.pie(top_10_sales['Sales'], 
        labels=top_10_sales['Country'],
        wedgeprops= {'edgecolor': 'black'},
        startangle=90,
        shadow= True,
        radius= 1.1, 
        counterclock = False,
        autopct='%1.1f%%' 
       )
plt.show()
```
